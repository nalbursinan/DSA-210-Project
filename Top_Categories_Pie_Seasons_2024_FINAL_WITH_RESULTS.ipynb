### 1. Load All Monthly Files and Add Season Column  
We load all monthly product-based Excel files and label each row with a corresponding season code.


```python
import pandas as pd
import glob

# Load all cleaned product-based Excel files
file_paths = sorted(glob.glob("UrunBazli_*_Temizlenmis_FinalTemiz.xlsx"))
all_data = []

# Map month codes to season labels
season_map = {
    "01": "K", "02": "K", "12": "K",
    "03": "I", "04": "I", "05": "I",
    "06": "Y", "07": "Y", "08": "Y",
    "09": "S", "10": "S", "11": "S"
}

# Read each file and assign season
for path in file_paths:
    df = pd.read_excel(path)
    month_code = path.split("_")[1][:2]
    df["MevsimKodu"] = season_map.get(month_code, "Unknown")
    all_data.append(df)

# Combine all into one DataFrame
all_months_df = pd.concat(all_data, ignore_index=True)
all_months_df.head()
```

# üìä Top Categories by Sales Quantity ‚Äì Seasonal Breakdown (2024)

This notebook shows pie charts of the top 10 best-selling categories for each season in 2024. All charts are based on net sales quantity.

```python

# üü¢ Kƒ±≈ü (Winter)

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

files = ['UrunBazli_0124_Temizlenmis_FinalTemiz.xlsx', 'UrunBazli_0224_Temizlenmis_FinalTemiz.xlsx', 'UrunBazli_1224_Temizlenmis_FinalTemiz.xlsx']
dfs = [pd.read_excel(f) for f in files]
df = pd.concat(dfs, ignore_index=True)

category_sales = df.groupby("Kategori")["Net Sats Adedi"].sum().sort_values(ascending=False)
top10 = category_sales.head(10)
other = category_sales.iloc[10:].sum()
if other > 0:
    top10["Other"] = other

colors = sns.color_palette("bright", len(top10))
plt.figure(figsize=(7, 7))
plt.pie(top10, labels=top10.index, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title("Top Categories by Sales Quantity ‚Äì Kƒ±≈ü (Winter)", fontsize=14)
plt.axis('equal')
plt.tight_layout()
plt.show()
```


**üìù Notes ‚Äì Winter:**
- During the winter season, categories related to occupational safety and protective gear stand out.
- High sales are observed in gloves, boots, and similar items.
- This may be due to the continuity of construction and outdoor work despite cold weather.
