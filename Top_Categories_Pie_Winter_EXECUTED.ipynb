```python

# Pie Chart: Top Categories by Sales Quantity  Winter Season (K)

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Combine data from January, February, December
files = [
    "UrunBazli_0124_Temizlenmis_FinalTemiz.xlsx",
    "UrunBazli_0224_Temizlenmis_FinalTemiz.xlsx",
    "UrunBazli_1224_Temizlenmis_FinalTemiz.xlsx"
]
dfs = [pd.read_excel(f) for f in files]
df = pd.concat(dfs, ignore_index=True)

# Aggregate by category
category_sales = df.groupby("Kategori")["Net Sats Adedi"].sum().sort_values(ascending=False)
top10 = category_sales.head(10)
other = category_sales.iloc[10:].sum()
if other > 0:
    top10["Other"] = other

# Pie chart
colors = sns.color_palette("bright", len(top10))
plt.figure(figsize=(7, 7))
plt.pie(top10, labels=top10.index, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title("Top Categories by Sales Quantity  Winter Season (K)", fontsize=14)
plt.axis('equal')
plt.tight_layout()
plt.show()
```
